# Analyzing the Categorical Variables from the Ames Housing Dataset

```python
# Create a new DataFrame with only the columns that are of categorical type
obj_df = df.select_dtypes(include='object')
```

```python
# extract the list of columns of this DataFrame
obj_cols = obj_df.columns
obj_cols
```

```python
# Create a function that takes a pandas DataFrame and a column name as input parameters
def describe_object(df, col_name):
    print(f'\nCOLUMN: {col_name}')
    print(f'{df[col_name].nunique()} different values')
    print(f'List of values: ')
    print(df[col_name].value_counts(dropna=False, normalize=True))
```

```python
describe_object(df, 'MSZoning')
```

```python
# Create a for loop that will call the created function for every element from the obj_cols list
for col_name in obj_cols:
    describe_object(df, col_name)
```

# Analyzing Numerical Variables from the Ames Housing Dataset

```python
# Create a new DataFrame with only the columns that are numerical
num_df = df.select_dtypes(include='number')
```

```python
# extract the list of columns of this DataFrame
num_cols = num_df.columns
num_cols
```

```python
# Create a function that takes a DataFrame and a column name as input parameters and describe them
def describe_numeric(df, col_name):
    print(f'\nCOLUMN: {col_name}')
    print(f'Minimum: {df[col_name].min()}')
    print(f'Maximun: {df[col_name].max()}')
    print(f'Average: {df[col_name].mean()}')
    print(f'Standard Deviation: {df[col_name].std()}')
    print(f'Median: {df[col_name].median()}')
```

```python
describe_numeric(df, 'SalePrice')
```

```python
# Create a for loop that will call the created function for every element from the num_cols list
for col_name in num_cols:
    describe_numeric(df, col_name)
```

# Visualizing the Ames Housing Dataset with Altair

```python
# import package
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# plot histogram of SalePrice variable
plt.figure()
plt.hist(df['SalePrice'])
plt.xlabel('Sale Price')
plt.ylabel('Counts of Records')
plt.show()
```

```python
# plot the histogram of LotArea variable
plt.figure()
plt.hist(df['LotArea'], bins=25)
plt.grid()
plt.xlabel('Lot Area')
plt.ylabel('Number os Records')
plt.show()
```

```python
# plot a scatter plot with LotArea as the x axis and SalePrice as the y axis 
plt.figure()
plt.scatter(df['LotArea'], df['SalePrice'])
plt.grid()
plt.xlabel('LotArea')
plt.ylabel('Sale Price')
plt.show()
```

```python
# plot the histogram for OverallCond
plt.figure()
plt.hist(df['OverallCond'])
plt.xlabel('OverallCond')
plt.ylabel('Numer of Records')
plt.show()
```

```python
# Build a boxplot with OverallCond on the x axis and SalePrice on the y axis
plt.figure(figsize=(10, 8))
sns.boxplot(x='OverallCond', y='SalePrice', data=df)
plt.show()
```

```python
# plot a bar chart for YrSold as its x axis and count() as its y axis
new_df = pd.DataFrame(df['YrSold'].value_counts())
new_df = new_df.reset_index().sort_values(by='index')

sns.barplot(x=new_df.iloc[:, 0], y=new_df.iloc[:, 1])
plt.show()
```

```python
# Plot a boxplot similar for YrSold as its x axis
new_df = pd.DataFrame()
new_df['YrSold'] = df['YrSold']
new_df['SalePrice'] = df['SalePrice']

plt.figure(figsize=(6, 10))
sns.boxplot(x='YrSold', y='SalePrice', data=new_df)
plt.show()
```

```python
# plot the histogram for OverallCond
new_df = pd.DataFrame(df['Neighborhood'].value_counts())
new_df = new_df.reset_index().sort_values(by='index')
plt.figure(figsize=(25, 8))
plt.bar(new_df.iloc[:, 0], new_df.iloc[:, 1])
plt.show()
```

```python
# analyze the relationship between SalePrice and Neighborhood by plotting a boxplot
new_df = pd.DataFrame(df[['Neighborhood', 'SalePrice']])
new_df = new_df.reset_index().sort_values(by='Neighborhood')

plt.figure(figsize=(20, 8))
sns.boxplot(x='Neighborhood', y='SalePrice', data=new_df)
plt.show()
```
