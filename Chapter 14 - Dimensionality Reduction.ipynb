# Loading and Cleaning the Dataset

```python
# import packages
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
# url path
url_path = 'https://raw.githubusercontent.com/PacktWorkshops/The-Data-Science-Workshop/master/Chapter14/Dataset/ad.data'
```

```python
# load the data
adData = pd.read_csv(url_path, header=None, error_bad_lines=False)
adData.head()
```

```python
# print the data shape
adData.shape
```

```python
# summary the numerical columns
adData.describe()
```

```python
# Separate the dependent and independent variables from our dataset
X = adData.loc[:, 0:1557]
y = adData[1558]
```

```python
# Print the first 15 examples of the independent variables
adData.head(15)
```

From the output, we can see that there are many missing values in the dataset, which are represented by ?.

```python
# print the data types of the dataset
adData.dtypes
```

The ? character made the first column be associated with object (string) data

```python
# Replace special characters with NaN values for the first four columns
for i in range(0, 3):
    X[i] = X[i].str.replace('?', 'nan').values.astype(float)
```

```python
X.head(15)
```

```python
# replace special characters for the integer feature
for i in range(3,1557):
  X[i] = X[i].replace("?", 'NaN').values.astype(float)
```

```python
# impute the mean of each column for the NaN values
for i in range(0, 1557):
    X[i] = X[i].fillna(X[i].mean())
```

```python
X.head(15)
```

```python
# scale the dataset using the minmaxScaler() function
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_trans = pd.DataFrame(scaler.fit_transform(X))
X_trans.head()
```

# Dimensionality Reduction Using Backward Feature Elimination

```python
# create a high-dimensional dataset
X_hd = pd.DataFrame(pd.np.tile(X_trans, (1, 2)))
X_hd.shape
```

```python
# Define the backward elimination model
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression

backModel = LogisticRegression()
rfe = RFE(backModel, 250)
```

```python
# Fit the backward elimination method to identify the best 250 features
import time

t0 = time.time()
rfe = rfe.fit(X_hd, y)
t1 = time.time()

print(f'Backward Elimination time: {round(t1-t0, 3)} s')
```

```python
# Display the features identified
rfe.get_support(indices=True)
```

```python
# split the dataset into training and testing sets
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_hd, y, test_size=0.3, random_state=123)

print(f'Training set shape: {X_train.shape}')
print(f'Test set shape: {X_test.shape}')
```

```python
# Transform the train and test sets
X_train_trans = rfe.transform(X_train)
X_test_trans = rfe.transform(X_test)

print(f'Training set shape: {X_train_trans.shape}')
print(f'Test set shape: {X_test_trans.shape}')
```

```python
# it a logistic regression model
import time

t0 = time.time()
RfeModel = LogisticRegression()
RfeModel.fit(X_train_trans, y_train)
t1 = time.time()

print(f'Time to fit the model: {round(t1-t0, 3)} s')
```

```python
# make predictions
pred = RfeModel.predict(X_test_trans)
```

```python
# calculate accuracy
print(f'Accuracy of Logistic Regression after backward elimination: {RfeModel.score(X_test_trans, y_test)}')
```

```python
# calculate confusion matrix
from sklearn.metrics import confusion_matrix

print(confusion_matrix(y_test, pred))
```

```python
# print classification report
from sklearn.metrics import classification_report

print(classification_report(y_test, pred))
```

# Dimensionality Reduction Using Forward Feature Selection

```python
# Create a high-dimensional dataset
X_hd = pd.DataFrame(pd.np.tile(X_trans, (1, 50)))
X_hd.shape
```

```python
# Split the high dimensional dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_hd, y, test_size=0.3, random_state=123)

print(f'Train shape before transformation: {X_train.shape}')
print(f'Test shape before transformation: {X_test.shape}')
```

```python
# define the threshold features
from sklearn.feature_selection import SelectKBest

feats = SelectKBest(k=250)
```

```python
# Iterate and get the best set of threshold features
t0 = time.time()
fit = feats.fit(X_train, y_train)
t1 = time.time()

print(f'Forward selction fit time: {round(t1-t0, 3)} s')
```

```python
# Create new training and test sets
features_train = fit.transform(X_train)
features_test = fit.transform(X_test)

print(f'Train shape after transformation: {features_train.shape}')
print(f'Test shape after transformation: {features_test.shape}')
```

```python
# fit a logistic regression model 
t0 = time.time()
forwardModel = LogisticRegression()
forwardModel.fit(features_train, y_train)
t1 = time.time()

print(f'Total training time: {round(t1-t0, 3)} s')
```

```python
# make prediction
pred = forwardModel.predict(features_test)
```

```python
# calculate accuracy
print(f'Accuracy of Logistic Regression on test set: {round(forwardModel.score(features_test, y_test), 2)}')
```

```python
# print confusion matrix
print(confusion_matrix(y_test, pred))
```

```python
# print classification report
print(classification_report(y_test, pred))
```

# Dimensionality Reduction Using PCA

```python
# fit the PCA function on the training set
from sklearn.decomposition import PCA

t0 = time.time()
pca = PCA().fit(X_train)
t1 = time.time()

print(f'PCA fitting time: {round(t1-t0, 3)} s')
```

```python
# determine the number of principal components by plotting the cumulative variance explained by all the principal components
import matplotlib.pyplot as plt

plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative explained variance')
plt.show()
```

From the plot, we can see that the first 250 principal components explain more than 90% of the variance. Based on this graph, we can decide how many principal components we want to have depending on the variability it explains. Let's select 250 components for fitting our model.

```python
# refit the training set for 250 components
pca = PCA(n_components=250).fit(X_train)
```

```python
# transform the training and test sets with the 250 principal components
X_pca = pca.transform(X_train)
X_test_pca = pca.transform(X_test)
```

```python
# verify the shapes of the train and test sets before transformation and after transformation
print(f'Training shape before transformation: {X_train.shape}')
print(f'Testing shape before transformation: {X_test.shape}')
print(f'Training shape after transformation: {X_pca.shape}')
print(f'Testting shape after transformation: {X_test_pca.shape}')
```

```python
# fit the logistic regression model
pcaModel = LogisticRegression()

t0 = time.time()
pcaModel.fit(X_pca, y_train)
t1 = time.time()

print(f'Total training time: {round(t1-t0, 3)} s')
```

```python
# make predictions
pred = pcaModel.predict(X_test_pca)
```

```python
# calculate accuracy
print(f'Accuracy of Logistic Regression model on test set: {pcaModel.score(X_test_pca, y_test)}')
```

```python
# print confusion matrix
print(confusion_matrix(y_test, pred))
```

```python
# print classification report
print(classification_report(y_test, pred))
```

```python
# calculate accuracy
print(f'Accuracy of Logistic Regression model: {icaModel.score(X_test_ica, y_test)}')
```

```python
# print confusion matrix
print(confusion_matrix(y_test, pred))
```
